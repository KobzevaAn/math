```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math as m
```

```python
cauchy = sts.cauchy(0, 1)
sample = cauchy.rvs(1000)
x = np.linspace(-5,5,200)

#построим гистограмму для полученной выборки

plt.hist(sample,bins=40, range = [-10,10], normed = True, histtype='stepfilled', alpha=1, label='Epdf')

#на этом же графике построим теоретическую функцуию плотность распределения

pdf = cauchy.pdf(x)
plt.plot(x, pdf, alpha=1, label='theoretical pdf')
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

Сгенерируем выборку объёма 1000 из стандартного распределения Коши

```python
cauchy = sts.cauchy(0, 1)
sample = cauchy.rvs(1000)
```

```python
average = 0
avr = []
n = np.linspace(50, 1000, 20)
value = [100, 1000, 10000]
for v in value:
    for nums in n:
        nums = int(nums)
        avr = []
        for i in range(v):
            sample = cauchy.rvs(nums)
            for l in sample:
                average += l
            average = average/nums
            avr.append(average)
        
#построим гистограмму выборки объемом 1000 для распредления среднего по выборке объема nums
        plt.hist(avr, bins = 30, range = [-10,10], normed = True, alpha = 1)
        plt.ylabel('f(x)')
        plt.xlabel('$x$')

#построим на каждом графике теоретическую функцию плотности нормального распредления Коши

        pdf = cauchy.pdf(x)
        plt.plot(x, pdf,  alpha = 2)
        plt.title('nums = %d, volume = '%nums + str(v))
        plt.savefig('graph/volume = %d, nums = %d.png'%(v, nums))
        plt.show()
```
