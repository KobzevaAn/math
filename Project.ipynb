```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math as m
```

```python
cauchy = sts.cauchy(0,1)
sample = cauchy.rvs(1000)
x = np.linspace(-5,5,200)

#построим гистограмму для полученной выборки

plt.hist(sample,bins=50, range = [-5,5], normed = True, histtype='stepfilled', alpha=1)

#на этом же графике построим теоретическую функцуию плотность распределения
pdf = cauchy.pdf(x)
plt.plot(x, pdf, alpha=2)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

Сгенерируем выборку объёма 1000 из стандартного распределения Коши

```python
cauchy = sts.cauchy(0, 1)
sample = cauchy.rvs(1000)
```

```python
average = 0
avr = []
n = [5, 10, 100]
for nums in n:
    for i in range(1000):
        sample = cauchy.rvs(nums)
        for l in sample:
            average += l
        average = average/nums
        avr.append(average)
#построим гистограмму выборки объемом 1000 для распредления среднего по выборке объема nums
    plt.hist(avr, bins = 40,range = [-10,10], label='nums = ' + str(nums), normed = True, alpha = 1)
    plt.ylabel('f(x)')
    plt.xlabel('$x$')
#построим на каждом графике теоретическую функцию плотности стандартного распредления Коши 
    pdf = cauchy.pdf(x)
    plt.plot(x, pdf,  alpha = 2)
    plt.legend(loc='upper right')
    plt.show()
```
